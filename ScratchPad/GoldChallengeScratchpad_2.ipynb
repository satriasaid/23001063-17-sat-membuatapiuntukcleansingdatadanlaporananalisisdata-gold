```python
import pandas as pd 
from sqlite3 import connect

conn = connect("postprocessed_tweet.db")
df = pd.read_sql('SELECT * FROM tweet_post',conn)
df.head()
```

```python
df_abusive = pd.read_csv("abusive.csv")
df_abusive.head()
```

```python
abusive = df_abusive['ABUSIVE'].to_list()
tweet = df["Tweet"].to_list()
```

```python
countAbuseList = []
countAbuse = 0
start_index = 0

tweet_join = ' '.join(tweet)

for l in range(len(abusive)):
    for i in range(len(tweet)):
        j = tweet_join.find(abusive[l], start_index)
        if(j!=-1):
            start_index = j+1
            countAbuse += 1
    l += 1
    countAbuseList.append(countAbuse)
    countAbuse = 0
    start_index = 0

df_abusive['COUNTER'] = countAbuseList
df_abusive.head()
df_abusive.to_csv('abuse_counted.csv')
```

```python
quoteList = ['let,let,lit,lot','lot,let','sandy,sander']
finding = ['let','lot','sand']
countfind = []
count = 0
start_index = 0

quote_join = ' '.join(quoteList)

for l in range(len(finding)):
    for i in range(len(quote_join)):
        j = quote_join.find(finding[l], start_index)
        if(j!=-1):
            start_index = j+1
            count += 1
    l += 1
    countfind.append(count)
    count = 0
    start_index = 0

print(countfind)
```
